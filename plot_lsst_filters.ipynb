```python
import numpy as np
import matplotlib.pylab as plt
```

```python
path = '/home/martin/auxtel_lsst/github/throughputs/baseline/'
file_label = 'filter_{0}.dat'
outdir = '/home/martin/Documents/IJCLab/presentaciones/a2c_pole_2022/'
```

```python
filters = ['u','g','r','i','z','y']
filter_colors = {'u':'b','g':'g','r':'r','i':'orange','z':'k','y':'grey'}
```

```python
fig=plt.figure(figsize=(10,6))
for f in filters:
    filter = np.loadtxt(path+file_label.format(f))
    l = filter[:,0]*10.
    val = filter[:,1]
    if f=='y':
        label = 'Y'
    else:
        label = f
    plt.plot(l,val,color=filter_colors[f],label=label)
    plt.fill_between(l,val,color=filter_colors[f],alpha=0.2)
plt.grid()
plt.xlim((2000,12000))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('$\\lambda \, [$'+'angstrom]',fontsize=15)
plt.ylabel('Filter transmission',fontsize=15)
plt.legend(fontsize=12)
plt.savefig(outdir+'filters.png')
```
